```
aws ec2 create-volume --generate-cli-skeleton > myspec.json

```  
  
   
![](images/skeleton.png) 
  
### Important   
  
This gives the opportunity to test before run.  
  
`"DryRun": true,` 
  
    
## Skeleton Execute. 
  
```
aws ec2 create-volume --cli-input-json file://volcreate.json
```
